```python
import pandas as pd
import numpy as np 
import networkx as nx 
```

```python
df= pd.read_csv('almacenes_entregas.csv', index_col=None)
```

```python
ALMACENES_ENTREGA=nx.from_pandas_edgelist(df,source='Almacenes',target='Puntos de Entrega',edge_attr='Trayectoria')
```
