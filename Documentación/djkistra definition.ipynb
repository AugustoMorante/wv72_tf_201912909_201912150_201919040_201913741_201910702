**ALGORITMO DE DJKISTRA**

**Implementación en pseudocódigo**

DIJKSTRA (Grafo G, nodo_fuente s)

para u ∈ V\[G\] hacer

distancia\[u\] = INFINITO

padre\[u\] = NULL

visto\[u\] = false

distancia\[s\] = 0

adicionar (cola, (s, distancia\[s\]))

mientras que cola no es vacía hacer

u = extraer_mínimo(cola)

visto\[u\] = true

para todos v ∈ adyacencia\[u\] hacer

si ¬ visto\[v\]

si distancia\[v\] \> distancia\[u\] + peso (u, v) hacer

distancia\[v\] = distancia\[u\] + peso (u, v)

padre\[v\] = u

adicionar(cola,(v, distancia\[v\]))

**Posible orden de complejidad**

Djkstra puede implementarse usando o no una cola de prioridad por lo
cual se tiene dos posibles respuestas:

UTILIZANDO COLA DE PRIORIDAD

$$O(|E| + |V|\log|V|)$$

SIN UTILIZAR COLA DE PRIORIDAD

$$O(|V|^{2} + E)$$
