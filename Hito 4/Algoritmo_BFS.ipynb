## Algoritmo BFS

```python
import grafo_hito4 as h4
```

```python
def quitarPeso(G):
  L = []
  PL = []
  for i, edges in enumerate(G):
    for j in edges:
      PL.append(j[0])
    L.append(PL)
    PL = []
  return L
```

```python
def bfs(G, s):
  n = len(G)
  visited = [False]*n
  parent = [None]*n
  queue = [s]
  visited[s] = True

  while queue:
    u = queue.pop(0)
    for v in G[u]:
      if not visited[v]:
        visited[v] = True
        parent[v] = u
        queue.append(v)

  return parent
```

```python
G = h4.G
G = quitarPeso(G)
for i, edges in enumerate(G):
  print(f"{i:2}: {edges}")
```

```python
path = bfs(G,0)
print(path)
```
